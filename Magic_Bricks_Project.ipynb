```python
# Filling the Null Values in DAta.

df.Bathroom = df.Bathroom.fillna(2)
df.Parking = df.Parking.fillna(1)
df.Type = df.Type.fillna('Unknown')
df.Per_Sqft = df.Per_Sqft.fillna(df.Per_Sqft.median())
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Removing OUTLIERS From the DAta.
# Step-1 Checking Outliers With Boxplot
sns.boxplot(df.Price,orient='h')
```

```python
# Step-2 Finding Q1 and Q3

Q1,Q3 = df.Price.quantile([0.25,0.75])
```

```python
Q1,Q3
```

```python
# Step-3 Finding IQR

IQR = Q3 -Q1
```

```python
IQR
```

```python
# STep-4 Finding UL and LL.

UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
# Step-5 Replacing Outliers with (UL).

df.Price = np.where(df.Price>UL,UL,df.Price)
```

```python
sns.boxplot(df.Price,orient='h')
```
